```python
def url_to_transcript(url):
    '''Returns transcript data specifically from scrapsfromtheloft.com.'''
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find_all('p')]
    print(url)
    return text
```

# Scrapping Beige Book

```python
import pandas as pd
import numpy as np
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, ElementNotVisibleException
from selenium.common.exceptions import TimeoutException
from selenium import webdriver


#webdriver links up with the browser and perform the actions like scraping

PATH = "C:\Program Files (x86)\chromedriver.exe"  
#the path where your chrome driver is in

driver = webdriver.Chrome(PATH) 
#browser we want to use is chrome and the location of it is in path

driver.get("https://www.federalreserve.gov/monetarypolicy/beige-book-archive.htm")
#opening up the website that we want to go to

delay = 15 # seconds
try:
    myElem = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, "/html/body/div[3]/div[2]/div[1]/div/div/div/ul/li[1]/a")))
    print ("Page is ready!")
except TimeoutException:
    print ("Loading took too much time!")


url=[]
for i in np.arange(1,27):
    url.append(driver.find_element(By.XPATH, f"/html/body/div[3]/div[2]/div[1]/div/div/div/ul/li[{i}]/a").get_attribute("href"))

#The following steps are done beacuse for 1996 and 2003, 
#they do not contain similar amount of beige book reports as other years
url_1996="https://www.federalreserve.gov/monetarypolicy/beigebook1996.htm"
url_2003="https://www.federalreserve.gov/monetarypolicy/beigebook2003.htm"
url.remove(url_1996)
url.remove(url_2003)

driver.quit()

url1=[]

driver = webdriver.Chrome(PATH)
def get_scarping(link):
    driver.get(link)
    for i in np.arange(1,9):
        url1.append(driver.find_element(By.XPATH, f"/html/body/div[3]/div[2]/div/table/tbody/tr[{i}]/td[2]/a[1]").get_attribute("href"))
    print(driver.current_url)
    return driver.current_url 

scrapings = []
for link in url:
    scrapings.append(get_scarping(link))
    
def get_scarping_1996(link):
    driver.get(link)
    for i in np.arange(1,3):
        url1.append(driver.find_element(By.XPATH, f"/html/body/div[3]/div[2]/div/table/tbody/tr[{i}]/td[2]/a[1]").get_attribute("href"))
    print(driver.current_url)
    return driver.current_url

def get_scarping_2003(link):
    driver.get(link)
    for i in np.arange(1,8):
        url1.append(driver.find_element(By.XPATH, f"/html/body/div[3]/div[2]/div/table/tbody/tr[{i}]/td[2]/a").get_attribute("href"))
    print(driver.current_url)
    return driver.current_url 

scrapings.append(get_scarping_1996(url_1996))
scrapings.append(get_scarping_2003(url_2003))
driver.quit()


text=[]
date=[]
driver = webdriver.Chrome(PATH)
for i in url1[0:40]:
    driver.get(i)
    text.append(driver.find_element(By.XPATH, "/html/body/div[3]/div[2]").text)
    date.append(driver.find_element(By.XPATH, "/html/body/div[3]/div[1]/div/div/div[1]/h2").text)
    
for i in url1[40:88]:
    driver.get(i)
    text.append(driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[2]/div[4]/div[1]").text)
    date.append(driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[2]/h1").text)
    
for i in url1[88:]:
    driver.get(i)
    text.append(driver.find_element(By.XPATH, "/html/body/p[1]/table/tbody/tr[2]/td[2]").text)
    date.append(driver.find_element(By.XPATH, "/html/body/p[1]/table/tbody/tr[1]/td[2]/strong/font").text)
    
driver.quit()
    
```

```python
df = pd.DataFrame({'date': date,'Beige_Book':text})
df['date'] = df['date'].str.replace('Beige Book -', '')
df['date'] = df['date'].str.replace(',', '')
df['date']=df['date'].astype(str)
df["date"]=pd.to_datetime(df["date"])
Beige_data=df.set_index(df.columns[0])
Beige_data
```

```python
Beige_data.to_csv("Beige_data.csv")
```
